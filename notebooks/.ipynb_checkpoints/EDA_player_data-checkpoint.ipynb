```python
from sklearn import preprocessing
y = df['PLUS_MINUS']
X = df.loc[:, ['START_POSITION','MIN','FGM','FGA','FG_PCT','FG3M','FG3A','FG3_PCT','FTM','FTA','FT_PCT','OREB','DREB',
           'REB','AST','STL','BLK','TO','PF','PTS']]
X = preprocessing.StandardScaler().fit_transform(X)
X = sm.add_constant(X)
m = sm.OLS(y, X).fit()
print(m.summary())
```

```python
y_point = df['PTS']
X = df.loc[:, ['START_POSITION','MIN','FGM','FGA','FG_PCT','FG3M','FG3A','FG3_PCT','FTM','FTA','FT_PCT','OREB','DREB',
           'REB','AST','STL','BLK','TO','PF','PLUS_MINUS']]
X = sm.add_constant(X)
m = sm.OLS(y_point, X).fit()
print(m.summary())
```

```python
# investigate relationship between pre season, regular season, and post season.
    # create a new column season
def season_creation(integer):
    if str(integer)[0] == '1':
        return 1
    elif str(integer)[0]  == '2':
        return 2
    else:
        return 3
        
df['season'] = df['GAME_ID'].apply(season_creation)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(5, 8))
sns.violinplot(data=df, x = 'season', y = 'PTS');
```

```python
X = df['season']
X = sm.add_constant(X)
m = sm.OLS(y, X).fit()
print(m.summary())
```

```python
# two sample t-test
import scipy.stats as stats
pre_season = df[df['season'] == 1]['PLUS_MINUS']
regular_season = df[df['season'] == 2]['PLUS_MINUS']
stats.ttest_ind(a=pre_season, b=regular_season, equal_var=True)
```

```python
post_season = df[df['season'] == 3]['PLUS_MINUS']
stats.ttest_ind(a=post_season, b=regular_season, equal_var=True)
```

```python
# explore time effect on player data
df_game = pd.read_csv('../data/raw/game_detials_0422/games.csv')
df_game = df_game[['GAME_DATE_EST', 'GAME_ID', 'HOME_TEAM_ID']]
df_new = df.merge(df_game, how='left', on='GAME_ID')
df_new
```

```python
df_new['loc'] = df_new.apply(lambda row: 1 if row['TEAM_ID'] == row['HOME_TEAM_ID'] else 0, axis=1)
df_new
```

```python
df_new['year'] = df_new.apply(lambda row: row['GAME_DATE_EST'].split('-')[0], axis=1)
df_new['month'] = df_new.apply(lambda row: row['GAME_DATE_EST'].split('-')[1], axis=1)
df_new['day'] = df_new.apply(lambda row: row['GAME_DATE_EST'].split('-')[2], axis=1)
df_new
```
