```python
df_all = pd.read_csv('../data/processing/game_features_after_DEA.csv')
df_all
```

```python
cor_matrix=df_all.corr()
cor_matrix.to_csv('cor_matrix_team_all_features.csv')
# we found that for point spread prediction, 
    # oppo_let : drop FTM, FTA, OREB
    # lags sub-section, only pointspread lags and PLUS_MINUS lags have correlation over 0.1 because they are highly correlated. so keep pointspread and its loc would be enough.
    # rol_avg: remove FTM, FTA, TO, PF; 73 best
    # vs_oppo: keep pointspread, loc, FIC
# for PTS prediction,
    # oppo_let remove FTM, FTA, OREB
    # lags remove FTM, FTA, OREB, STL, BLK, TO, PF, pointspread, PLUS_MINUS
    # rol_avg: remove FTM, FTA, TO, PF; 73 best
    # vs_oppo: remove FTM, FTA, OREB, STL, BLK, TO, PF, pointspread, PLUS_MINUS
```

```python
# analysis of year's effect on relationships between target variables and features
years = ['2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
```

```python
# correlation change over time pointspread
pd.set_option('display.max_rows', 1000)
params_list = []
for year in years:
    df_year = df_all[df_all['GAME_DATE_EST'].str.contains(year)]
    params_list.append(df_year[df_year.columns[1:]].corr()['pointspread'][:].round(2))
df_params = pd.concat(params_list, axis=1)
df_params # no need to drop
```

```python
# correlation change over time PTS_home
pd.set_option('display.max_rows', 1000)
params_list = []
for year in years:
    df_year = df_all[df_all['GAME_DATE_EST'].str.contains(year)]
    params_list.append(df_year[df_year.columns[1:]].corr()['PTS_team'][:].round(2))
df_params = pd.concat(params_list, axis=1)
df_params # no need to drop
```

```python
# correlation change with different loc pointspread
pd.set_option('display.max_rows', 1000)
locs = [0, 1]
params_list = []
for loc in locs:
    df_loc = df_all[df_all['loc'] == loc]
    params_list.append(df_loc[df_loc.columns[1:]].corr()['pointspread'][:])
df_params = pd.concat(params_list, axis=1)
df_params # there is a difference
```

```python
# correlation change with different loc PTS
pd.set_option('display.max_rows', 1000)
locs = [0, 1]
params_list = []
for loc in locs:
    df_loc = df_all[df_all['loc'] == loc]
    params_list.append(df_loc[df_loc.columns[1:]].corr()['PTS_team'][:].round(2))
df_params = pd.concat(params_list, axis=1)
df_params # no difference
```

```python
# correlation change with different season type pointspread
pd.set_option('display.max_rows', 1000)
params_list = []
df_all = df_all.astype({'GAME_ID':str})
df_pre = df_all[df_all['GAME_ID'].str.startswith('1')]
df_reg = df_all[df_all['GAME_ID'].str.startswith('2')]
df_pos = df_all[df_all['GAME_ID'].str.startswith(('4', '5'))]
params_list.append(df_pre[df_pre.columns[1:]].corr()['pointspread'][:].round(2))
params_list.append(df_reg[df_reg.columns[1:]].corr()['pointspread'][:].round(2))
params_list.append(df_pos[df_pos.columns[1:]].corr()['pointspread'][:].round(2))
df_params = pd.concat(params_list, axis=1)
df_params # drop pre season
```

```python
# correlation change with different season type PTS_team
pd.set_option('display.max_rows', 1000)
params_list = []

df_pre = df_all[df_all['GAME_ID'].str.startswith('1')]
df_reg = df_all[df_all['GAME_ID'].str.startswith('2')]
df_pos = df_all[df_all['GAME_ID'].str.startswith(('4', '5'))]
params_list.append(df_pre[df_pre.columns[1:]].corr()['PTS_team'][:].round(2))
params_list.append(df_reg[df_reg.columns[1:]].corr()['PTS_team'][:].round(2))
params_list.append(df_pos[df_pos.columns[1:]].corr()['PTS_team'][:].round(2))
df_params = pd.concat(params_list, axis=1)
df_params # drop pre season
```

```python
y = df_all['pointspread']
X = df_all.iloc[:, 53:]
X['loc'] = df_all['loc']
X
```

```python
feature_names = df_all.columns[53:]
feature_names
```

```python
y = df_all['pointspread']
X = df_all.iloc[:, 53:]
X['loc'] = df_all['loc']
# Create a random forest regressor
regr = RandomForestRegressor()

# Fit the model to the data
regr.fit(X, y)

# Get the feature importance scores
importances = regr.feature_importances_

# Zip the feature names together with the importance scores and sort them by importance
feature_importances = zip(feature_names, importances)

# Print the feature importance scores
for feature, importance in feature_importances:
    print(feature, importance)

# we found that for point spread prediction, 
    # oppo_let : (correlation <0.1) drop FTM, FTA, OREB
        # (random forest <0.001) EFG%, PPS, PTS, FG3A, FG3M, FGM, FGA, FIC
    # lags sub-section, only pointspread lags and PLUS_MINUS lags have correlation over 0.1 because they are highly correlated. so keep pointspread and its loc would be enough.
        # FG3M, FG3M, FGM, FTM, EFG%, STL, PPS, FG3A, TO, PLUS/pointspread > 7, PF, AST, DREB, OREB, BLK
    # rol_avg: remove FTM, FTA, TO, PF; 73 best
        # PPS, FG3M, EFG%, FG3A, PTS, FGM, FGA, FIC, DREB, FTA, REB, PF
    # vs_oppo: keep pointspread, loc, FIC
        # FG3M, BLK, EFG%, FGM, FG3A, OREB, DREB, AST, FGA, FG3A, STL, TO, REB, PF
```

```python
y_pts = df_all['PTS_team']
# Create a random forest regressor
regr_pts = RandomForestRegressor()

# Fit the model to the data
regr_pts.fit(X, y_pts)

# Get the feature importance scores
importances = regr_pts.feature_importances_

# Zip the feature names together with the importance scores and sort them by importance
feature_importances = zip(feature_names, importances)

# Print the feature importance scores
for feature, importance in feature_importances:
    print(feature, importance)
# for PTS prediction,
    # (correlation < 0.1) oppo_let remove FTM, FTA, OREB
        # (feature importance < 0.001) pointspread, PLUS, FG3M, EFG%, PPS, FIC
    # lags remove FTM, FTA, OREB, STL, BLK, TO, PF, pointspread, PLUS_MINUS
        # pointspread, AST, REB, FGM, FG3M, FG3A, FTM, OREB, DREB, STL, BLK, TO, PF, PLUS, EFG%, PPS 
    # rol_avg: remove FTM, FTA, TO, PF; 82 best
        # pointspread, FG, REB, FGA, FG3M, FG3A, FTM, FTA, DREB, PLUS, EFG%, PPS, FIC
    # vs_oppo: remove FTM, FTA, OREB, STL, BLK, TO, PF, pointspread, PLUS_MINUS
        # pointspread, FG, AST, FGM, FG3M, FG3A, FTM, FTA, OREB, DREB, STL, BLK, TO, PF, PLUS, EFG%, PPS
```
